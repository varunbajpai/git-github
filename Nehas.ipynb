```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
```

```python
stock_name = 'MSFT'                   #stock code for microsoft
start_date = '2010-01-01'
end_date = '2016-12-31'
panel_data = data.DataReader('MSFT', 'yahoo', start_date, end_date)
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.grid(color='grey', linestyle='-', linewidth=2)
ax.plot(panel_data['High'], label='MSFT')
ax.set_xlabel('Date')
ax.set_ylabel('Price Per Day')
ax.legend()
```
